```python
from skimage import io
from skimage import color
import skimage.filter as filters
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np 
% matplotlib inline
```

```python
ariel = io.imread('http://i.imgur.com/z99OZND.jpg')
snow_white = io.imread('http://i.imgur.com/h2JU5P0.jpg')
lion_king = io.imread('http://i.imgur.com/fHroiTd.jpg')
```

```python
def hsv_decomposition(image):
    hsv = color.rgb2hsv(image)
    h = hsv[:, :, 0]
    s = hsv[:, :, 1]
    v = hsv[:, :, 2]
    plt.figure(figsize = (10, 10))
    plt.subplot(221) 
    plt.title('Original image')
    plt.axis('off')
    plt.imshow(image)
    plt.subplot(222)
    plt.title('Hue')
    plt.gca().yaxis.set_major_locator(MaxNLocator(prune = 'lower'))
    plt.hist(h.flatten(), bins = 50, color = 'red')
    plt.subplot(223)    
    plt.title('Saturation')
    plt.gca().yaxis.set_major_locator(MaxNLocator(prune = 'lower'))
    plt.hist(s.flatten(), bins = 50, color = 'green')
    plt.subplot(224) 
    plt.title('Value')
    plt.gca().yaxis.set_major_locator(MaxNLocator(prune = 'lower'))
    plt.hist(v.flatten(), bins = 50, color = 'blue')
    plt.show()
```

```python
hsv_decomposition(ariel)
```

```python
def saturation_filter(image):
    hsv = color.rgb2hsv(image)
    max_value = 0.3
    hsv[:, :, 1][hsv[:, :, 1] > max_value] = max_value
    new_image = color.hsv2rgb(hsv)
    plt.figure(figsize = (10, 10))
    plt.subplot(121)
    plt.axis('off')
    plt.title('Original image')
    plt.imshow(image)
    plt.subplot(122)
    plt.axis('off')
    plt.title('Saturated image')
    plt.imshow(new_image)
```

```python
saturation_filter(ariel)
```

```python
def brightness_filter(image):
    hsv = color.rgb2hsv(image)
    start_value = 0
    step = 0.05
    for i in range(20):
        hsv[:, :, 2][np.logical_and(hsv[:, :, 2] >= start_value, hsv[:, :, 2] < start_value + step)] = start_value + step
        start_value += step   
    new_image = color.hsv2rgb(hsv)
    plt.figure(figsize = (10, 10))
    plt.subplot(121)
    plt.axis('off')
    plt.title('Original image')
    plt.imshow(image)
    plt.subplot(122)
    plt.axis('off')
    plt.title('Bright image')
    plt.imshow(new_image)
```

```python
brightness_filter(ariel)
```

```python
saturation_filter(snow_white)
```

```python
saturation_filter(lion_king)
```

```python
brightness_filter(snow_white)
```

```python
brightness_filter(lion_king)
```
